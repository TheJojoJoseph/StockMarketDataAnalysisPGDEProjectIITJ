# 2. Choose any ten assets from the market

> Add blockquote



We shall select 10 random tickers from the dataset that this project will analyse.

```python
import os
import numpy as np


companyList = os.listdir(path+"/comp_stock_data/stock_data_NSE/")

random10Picks = np.random.choice(companyList, 10)
print("---Selected Stocks---")
for index, value  in enumerate(random10Picks):
    print(f"{index+1}\t{value}")
```

## 2.1. Creating Data Path
We shall create the path strings to the relevant dataset from 2023 to 2021

```python
#creating paths to the stock files
import json

paths={}
#generate paths for random10Picks
def generatePath(company):
  for year in range(2023, 2021, -1):
    paths[company] = path+"/comp_stock_data/stock_data_NSE/"+company+"/"+company+"_"+str(year)+".csv"

for company in random10Picks:
  generatePath(company)

print(json.dumps(paths,indent=4))
```

#  3. Gather the closing prices for your chosen assets over the last 3 years

We shall create a python object that holds the data for last 3 years from all the 10 dataframes.

```python
import pandas as pd

dataframes = {}
for company in random10Picks:
  dataframe = pd.read_csv(paths[company])
  if dataframe.empty == False:
    dataframes[company] = dataframe


print(dataframes)
```

# 4. Draw the time graph and Histogram of returns of each individual asset. Obtain the measures of descriptive statistics

We shall attempt to draw the following for each individual asset:
1. Time graph
2. Histogram

```python
import matplotlib.pyplot as plt
import seaborn as sns

for company in random10Picks:
    try:
        df = dataframes[company]


        # Calculating returns
        df['Returns'] = df['Close'].pct_change()
        df = df.dropna() #drop nan rows

        # Time graph of closing prices
        plt.figure(figsize=(10, 5))
        plt.plot(df['Date'], df['Close'])
        plt.xlabel('Date')
        plt.ylabel('Closing Price')
        plt.title(f'Closing Prices for {company}')
        plt.grid(True)
        plt.show()

        # Histogram of returns
        plt.figure(figsize=(10, 5))
        sns.histplot(df['Returns'], kde=True)
        plt.xlabel('Returns')
        plt.ylabel('Frequency')
        plt.title(f'Distribution of Returns for {company}')
        plt.grid(True)
        plt.show()

        # Descriptive statistics
        print(f"Descriptive Statistics for {company}:")
        print(df['Returns'].describe())
        print("\n")

    except Exception as e:
        print(f"Error processing {company}: {e}")
```

# 5. Correlation Analysis of the Returns

We shall attempt to do some correlation analysis and understand the Returns data

```python

returns_data = pd.DataFrame()

for company in random10Picks:
    try:
        df = dataframes[company]
        df['Returns'] = df['Close'].pct_change()
        df = df.dropna()
        returns_data[company] = df['Returns']
    except Exception as e:
        print(f"Error processing {company} for correlation: {e}")

# Calculating the correlation matrix
correlation_matrix = returns_data.corr()

# Plotting the correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Stock Returns')
plt.show()

# Print the correlation matrix (optional)
correlation_matrix
```

# 6. Come up with a combined index which is the weighted average of all the picked assets

```python
weights_list = [1/len(random10Picks)] * len(random10Picks)  # Equal weights

combined_index = pd.DataFrame()
for i, company in enumerate(random10Picks):
    try:
        df = dataframes[company]
        df['Returns'] = df['Close'].pct_change()


        # Apply the weight to the returns
        df['Weighted_Returns'] = df['Returns'] * weights_list[i]

        if combined_index.empty:
            combined_index = df[['Date', 'Weighted_Returns']].copy()
            combined_index = combined_index.rename(columns={'Weighted_Returns': company})
        else:
            combined_index[company] = df['Weighted_Returns']

    except Exception as e:
        print(f"Error processing {company} for combined index: {e}")

# Calculate the weighted average of returns
combined_index['Combined_Index'] = combined_index.iloc[:, 1:].sum(axis=1)

# Display or further analyze the combined index
print(combined_index)

# Plot the combined index over time
plt.figure(figsize=(12, 6))
plt.plot(combined_index['Date'], combined_index['Combined_Index'])
plt.xlabel('Date')
plt.ylabel('Combined Index Value')
plt.title('Combined Index of Selected Assets')
plt.grid(True)
plt.show()
```

# 7. Present the Markowitz curve, CML and SML for a security that you may like out of 10 assets selected.

## Markowitz Curve and CML

We are generating 1500 portfolios of 10 random stocks and plotting it to see which of them have more return to risk ratio

### CML
Shows the relationship between risk and return for a portfolio. CML is a straight line that connects a risk-free rate with a market portfolio on the Efficient Frontier. CML is more realistic and empirical, and only plots portfolios that are well-diversified


```python
num_portfolios = 1500
num_assets = len(returns_data.columns)

portfolio_data = pd.DataFrame(columns=['Returns', 'Volatility'])

for _ in range(num_portfolios):
    # Generating random weights for the assets, ensuring the sum is 1
    weights = np.random.random(num_assets)
    weights /= np.sum(weights)

    # Calculating the portfolio return
    portfolio_return = np.sum(returns_data.mean() * weights) * 252 # annualize return

    # Calculate the portfolio volatility (standard deviation)
    portfolio_std_dev = np.sqrt(np.dot(weights.T, np.dot(returns_data.cov() * 252, weights))) # annualize volatility


    # Store the portfolio return and volatility
    portfolio_data = pd.concat([portfolio_data, pd.DataFrame({'Returns': portfolio_return, 'Volatility': portfolio_std_dev}, index=[0])], ignore_index=True)

# Plotting
portfolio_data['Sharpe_Ratio'] = portfolio_data['Returns'] / portfolio_data['Volatility'] #Sharpe Ratio

# Find the portfolio with the highest Sharpe Ratio
best_portfolio = portfolio_data.loc[portfolio_data['Sharpe_Ratio'].idxmax()]

# Assuming we have a risk-free rate (e.g., from a government bond yield)
risk_free_rate = 0.02  # Example: 2% risk-free rate

# Calculate the Sharpe Ratio of the best portfolio
sharpe_ratio = best_portfolio['Sharpe_Ratio']

cml_intercept = risk_free_rate

cml_slope = sharpe_ratio

# Generate points for the CML line
volatilities = np.linspace(0, portfolio_data['Volatility'].max(), 100)
expected_returns_cml = cml_intercept + cml_slope * volatilities

# Plot the Markowitz efficient frontier with Sharpe Ratio
plt.figure(figsize=(10, 6))
plt.scatter(portfolio_data['Volatility'], portfolio_data['Returns'], marker='o', s=10, alpha=0.7, label='Portfolios')
plt.scatter(best_portfolio['Volatility'], best_portfolio['Returns'], marker='*', s=200, color='red', label='Highest Sharpe Ratio Portfolio')
plt.plot(volatilities, expected_returns_cml, color='blue', linestyle='-', linewidth=2, label='Capital Market Line (CML)')

plt.xlabel('Volatility (Standard Deviation)')
plt.ylabel('Expected Return')
plt.title('Markowitz Efficient Frontier with Highest Sharpe Ratio Portfolio')
plt.grid(True)

# Annotate the best portfolio with its Sharpe Ratio
plt.annotate(f"Sharpe Ratio: {best_portfolio['Sharpe_Ratio']:.2f}",
             (best_portfolio['Volatility'], best_portfolio['Returns']),
             textcoords="offset points", xytext=(5,10), ha='left')


plt.legend()
plt.show()

#print the companies in this portfolip
print("Portfolio Companies:")
for i, company in enumerate(returns_data.columns):
    if weights[i] > 0: # Only print companies with non-zero weights
        print(f"- {company}: Weight = {weights[i]:.4f}")
```

## SML
Shows the risk and return of the market at a given time, and the expected returns of individual assets. SML plots all portfolios, including inefficient ones. SML is more theoretical

```python
selected_security = random10Picks[0]

try:
    security_data = dataframes[selected_security]
    security_data['Returns'] = security_data['Close'].pct_change()
    security_data = security_data.dropna()

    # 'combined_index' is market portfolio returns

    merged_data = pd.merge(security_data[['Date', 'Returns']], combined_index[['Date', 'Combined_Index']], on='Date', how='inner')
    merged_data.dropna(inplace=True)

    # Calculate the beta
    beta = np.cov(merged_data['Returns'], merged_data['Combined_Index'])[0, 1] / np.var(merged_data['Combined_Index'])

    # Security Market Line equation: E(Ri) = Rf + βi * (E(Rm) - Rf)
    # E(Rm) is the expected return of the market (the combined index in this case)
    expected_market_return = combined_index['Combined_Index'].mean() * 252  # Annualized
    expected_return_security = risk_free_rate + beta * (expected_market_return - risk_free_rate)


    # Plotting the SML
    plt.figure(figsize=(8, 6))
    plt.scatter(beta, expected_return_security, color='red', label=selected_security, s=100)  # Plot the selected security's beta and expected return

    #Generate points for the SML
    betas = np.linspace(0, 2, 100) #range of beta values
    expected_returns_sml = risk_free_rate + betas * (expected_market_return-risk_free_rate)
    plt.plot(betas, expected_returns_sml, color='blue', linestyle='-', linewidth=2, label='Security Market Line (SML)')

    plt.xlabel('Beta')
    plt.ylabel('Expected Return')
    plt.title('Security Market Line')
    plt.grid(True)
    plt.legend()
    plt.show()

    print(f"Beta of {selected_security}: {beta:.4f}")
    print(f"Expected Return of {selected_security}: {expected_return_security:.4f}")
except Exception as e:
    print(f"Error calculating SML for {selected_security}: {e}")
```

## 7.1 Obtain the return and the risk of combined index.


```python
# Calculating the return and risk of the combined index
combined_index_return = combined_index['Combined_Index'].mean() * 252  # Annualized return
combined_index_risk = combined_index['Combined_Index'].std() * np.sqrt(252)  # Annualized risk

# Eliminating invalid data (inf or nan)
if np.isfinite(combined_index_return) and np.isfinite(combined_index_risk):
    print(f"Combined Index Return: {combined_index_return:.4f}")
    print(f"Combined Index Risk: {combined_index_risk:.4f}")
else:
    print("Invalid data encountered (inf or nan). Cannot calculate return and risk.")
```
